# Problem 2

```python
# import
import mltools as ml
np.random.seed(0)  # set the random number seed at the beginning,to ensure consistent behavior each time
iris = np.genfromtxt("data/iris.txt", delimiter=None)  # load the text file
```

```python
# problem 2.1

Y = iris[:, -1]  # target value is the last column //[:,-1] means take every last elements of each row
X = iris[:, 0:2]  # features are the other columns //[row, column]

X, Y = ml.shuffleData(X, Y)

Xtr, Xva, Ytr, Yva = ml.splitData(X, Y, 0.75);

fig = plt.figure(1)
Position = range(1,5)
K = [1,5,10,50]
for c in range(0,4):
    ax = fig.add_subplot(2, 2, Position[c])
    ax.set_xlabel('K=' + str(K[c]))
    fig.set_size_inches(15,15)
    knn = ml.knn.knnClassify(Xtr, Ytr, K[c])  # training model
    YvaHat = knn.predict(Xva)  # make prediction
    ml.plotClassify2D(knn, Xtr, Ytr)
    
plt.show()
```

```python
# problem 2.2

Y = iris[:, -1]  # target value is the last column //[:,-1] means take every last elements of each row
X = iris[:, 0:2]  # features are the other columns //[row, column]
X, Y = ml.shuffleData(X, Y)
Xtr, Xva, Ytr, Yva = ml.splitData(X, Y, 0.75);

K = [1, 2, 5, 10, 50, 100, 200]
errVal = []
errTra = []
for i, K in enumerate(K):
    learner = ml.knn.knnClassify(Xtr, Ytr, K)
    Yhat = learner.predict(Xva)
    Yhat2 = learner.predict(Xtr)
    errVal.append(np.sum(Yhat != Yva)/np.size(Yva))
    errTra.append(np.sum(Yhat2!= Ytr)/np.size(Ytr))

Karr = (1,2,5,10,50,100,200)

print(errVal)
print(errTra)
plt.semilogx(Karr, errVal,'green')
plt.semilogx(Karr, errTra,'red')

plt.show()
```

Based on the error rates, I would recommend K to be 50 because it has the lowest error rate as K increases while the learner model would neither too complex or too simple.

```python
#problem 2.3

Y = iris[:, -1]
X = iris[:, 0:-1] #get all 4 features
X, Y = ml.shuffleData(X, Y)
Xtr, Xva, Ytr, Yva = ml.splitData(X, Y, 0.75);

K = [1, 2, 5, 10, 50, 100, 200]
errVal = []
errTra = []
for i, K in enumerate(K):
    learner = ml.knn.knnClassify(Xtr, Ytr, K)
    Yhat = learner.predict(Xva)
    Yhat2 = learner.predict(Xtr)
    errVal.append(np.sum(Yhat != Yva)/np.size(Yva))
    errTra.append(np.sum(Yhat2!= Ytr)/np.size(Ytr))

Karr = (1,2,5,10,50,100,200)

print(errVal)
print(errTra)
plt.semilogx(Karr, errVal,'green')
plt.semilogx(Karr, errTra,'red')

plt.show()
```

In this case, when apply all 4 features to knn learner, both the plots and recommendation will be different.
The recommendation will be K = 10 because it has the lowest error rate while the model is simpler than other Ks with the same error rate.
